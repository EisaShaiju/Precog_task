## LEXICAL RICHNESS

```python
# Import required libraries
import os
import json
import re
from pathlib import Path
from collections import Counter
import pandas as pd
import numpy as np

# Function to extract words from text
def extract_words(text):
    """Extract words from text (lowercase, alphanumeric only)"""
    words = re.findall(r'\b[a-z]+\b', text.lower())
    return words

# Function to calculate Type-Token Ratio (TTR)
def calculate_ttr(text):
    """
    Calculate Type-Token Ratio (TTR)
    TTR = Unique words / Total words
    """
    words = extract_words(text)
    if len(words) == 0:
        return 0
    unique_words = len(set(words))
    total_words = len(words)
    ttr = unique_words / total_words
    return ttr

# Function to calculate Hapax Legomena
def calculate_hapax(text, sample_size=5000):
    """
    Calculate Hapax Legomena count
    Hapax = Number of words that appear only once in a 5,000-word sample
    """
    words = extract_words(text)
    
    # Take only first 5000 words as sample
    sample_words = words[:sample_size]
    
    if len(sample_words) == 0:
        return 0
    
    # Count word frequencies
    word_counts = Counter(sample_words)
    
    # Count words that appear exactly once
    hapax_count = sum(1 for count in word_counts.values() if count == 1)
    
    return hapax_count

# Function to analyze lexical richness for a single text
def analyze_lexical_richness(text, label=""):
    """
    Analyze lexical richness metrics for a given text
    Returns a dictionary with TTR and Hapax counts
    """
    ttr = calculate_ttr(text)
    hapax = calculate_hapax(text, sample_size=5000)
    
    return {
        'label': label,
        'ttr': ttr,
        'hapax_legomena': hapax,
        'total_words': len(extract_words(text)),
        'unique_words': len(set(extract_words(text)))
    }

print("✓ Lexical Richness functions loaded successfully")
print("  - calculate_ttr(text)")
print("  - calculate_hapax(text, sample_size=5000)")
print("  - analyze_lexical_richness(text, label)")
```

## SYNTACTIC COMPLEXITY

```python
# Import required library for syntactic analysis
import spacy

# Load spaCy model (if not already loaded)
try:
    nlp = spacy.load("en_core_web_sm")
    print("✓ spaCy model loaded successfully")
except:
    print("⚠ Installing spaCy model...")
    import subprocess
    subprocess.run(["python", "-m", "spacy", "download", "en_core_web_sm"])
    nlp = spacy.load("en_core_web_sm")
    print("✓ spaCy model installed and loaded")

# Function to calculate average sentence length
def calculate_avg_sentence_length(text):
    """Calculate average number of words per sentence"""
    doc = nlp(text[:1000000])  # Limit to 1M chars to avoid memory issues
    sentences = list(doc.sents)
    if len(sentences) == 0:
        return 0
    total_words = sum(len([token for token in sent if not token.is_punct]) for sent in sentences)
    return total_words / len(sentences)

# Function to calculate dependency tree depth
def calculate_avg_dependency_depth(text, sample_size=100):
    """Calculate average dependency tree depth for a sample of sentences"""
    doc = nlp(text[:500000])  # Limit for performance
    sentences = list(doc.sents)[:sample_size]
    
    def get_tree_depth(token, depth=0):
        if not list(token.children):
            return depth
        return max(get_tree_depth(child, depth + 1) for child in token.children)
    
    depths = []
    for sent in sentences:
        if sent.root:
            depth = get_tree_depth(sent.root)
            depths.append(depth)
    
    return np.mean(depths) if depths else 0

# Function to calculate POS tag ratios
def calculate_pos_ratios(text):
    """Calculate ratios of different part-of-speech tags"""
    doc = nlp(text[:1000000])
    pos_counts = Counter([token.pos_ for token in doc if not token.is_punct and not token.is_space])
    total = sum(pos_counts.values())
    
    if total == 0:
        return {'noun_ratio': 0, 'verb_ratio': 0, 'adj_ratio': 0, 'adv_ratio': 0}
    
    return {
        'noun_ratio': (pos_counts['NOUN'] + pos_counts['PROPN']) / total,
        'verb_ratio': pos_counts['VERB'] / total,
        'adj_ratio': pos_counts['ADJ'] / total,
        'adv_ratio': pos_counts['ADV'] / total
    }

# Combined function
def analyze_syntactic_complexity(text, label=""):
    """Analyze syntactic complexity metrics"""
    avg_sent_len = calculate_avg_sentence_length(text)
    avg_depth = calculate_avg_dependency_depth(text, sample_size=100)
    pos_ratios = calculate_pos_ratios(text)
    
    return {
        'label': label,
        'avg_sentence_length': avg_sent_len,
        'avg_dependency_depth': avg_depth,
        **pos_ratios
    }

print("✓ Syntactic Complexity functions loaded successfully")
print("  - calculate_avg_sentence_length(text)")
print("  - calculate_avg_dependency_depth(text, sample_size=100)")
print("  - calculate_pos_ratios(text)")
print("  - analyze_syntactic_complexity(text, label)")
```

## POS DISTRIBUTION

```python
# Function to calculate POS distribution (counts and proportions for all POS tags)
def calculate_pos_distribution(text, label=""):
    """
    Calculate complete POS tag distribution with counts and proportions.
    Returns counts and proportions for all POS tags found in the text.
    """
    doc = nlp(text[:1000000])  # Limit to 1M chars to avoid memory issues
    pos_counts = Counter([token.pos_ for token in doc if not token.is_punct and not token.is_space])
    total = sum(pos_counts.values())
    
    result = {'label': label, 'total_pos_tokens': total}
    
    # Add counts and proportions for each POS tag
    for pos, count in pos_counts.items():
        result[f'count_{pos}'] = int(count)
        result[f'prop_{pos}'] = float(count / total) if total > 0 else 0.0
    
    return result

print("✓ POS Distribution function loaded successfully")
print("  - calculate_pos_distribution(text, label)")
```

## PUNCTUATION DENSITY

```python
# Import visualization library
import matplotlib.pyplot as plt
import seaborn as sns

# Function to calculate punctuation density
def calculate_punctuation_density(text, label=""):
    """Calculate density of different punctuation marks per 1000 words"""
    words = extract_words(text)
    total_words = len(words)
    
    if total_words == 0:
        return {'label': label, 'semicolon': 0, 'em_dash': 0, 'exclamation': 0, 'comma': 0, 'period': 0}
    
    # Count punctuation marks
    semicolons = text.count(';')
    em_dashes = text.count('—') + text.count('--')  # em-dash or double hyphen
    exclamations = text.count('!')
    commas = text.count(',')
    periods = text.count('.')
    
    # Calculate per 1000 words
    factor = 1000 / total_words
    
    return {
        'label': label,
        'semicolon': semicolons * factor,
        'em_dash': em_dashes * factor,
        'exclamation': exclamations * factor,
        'comma': commas * factor,
        'period': periods * factor
    }

print("✓ Punctuation Density function loaded successfully")
print("  - calculate_punctuation_density(text, label)")
```

### CLASS 1 - PUNCTUATION DENSITY

```python
# Analyze punctuation density for Class 1 novels
class1_punct_results = []

for novel_name, text in novels_class1:
    results = calculate_punctuation_density(text, label=f"{novel_name} (Class 1)")
    class1_punct_results.append(results)

class1_punct_df = pd.DataFrame(class1_punct_results)

# Create heatmap
plt.figure(figsize=(10, 6))
punct_data = class1_punct_df[['semicolon', 'em_dash', 'exclamation', 'comma', 'period']].values
sns.heatmap(punct_data, annot=True, fmt='.2f', cmap='YlOrRd', 
            xticklabels=['Semicolon', 'Em-dash', 'Exclamation', 'Comma', 'Period'],
            yticklabels=class1_punct_df['label'].values,
            cbar_kws={'label': 'Per 1000 words'})
plt.title('Class 1 - Punctuation Density Heatmap')
plt.tight_layout()
plt.show()

print(class1_punct_df.to_string(index=False))
```

```python
# Function to calculate Flesch-Kincaid Grade Level
def calculate_flesch_kincaid(text, label=""):
    """
    Calculate Flesch-Kincaid Grade Level
    Formula: 0.39 * (total words / total sentences) + 11.8 * (total syllables / total words) - 15.59
    """
    # Count sentences (approximate)
    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if s.strip()]
    total_sentences = len(sentences)
    
    if total_sentences == 0:
        return {'label': label, 'fk_grade_level': 0, 'avg_syllables_per_word': 0}
    
    # Count words
    words = extract_words(text)
    total_words = len(words)
    
    if total_words == 0:
        return {'label': label, 'fk_grade_level': 0, 'avg_syllables_per_word': 0}
    
    # Count syllables (approximate using vowel groups)
    def count_syllables(word):
        word = word.lower()
        vowels = 'aeiouy'
        syllable_count = 0
        previous_was_vowel = False
        
        for char in word:
            is_vowel = char in vowels
            if is_vowel and not previous_was_vowel:
                syllable_count += 1
            previous_was_vowel = is_vowel
        
        # Adjust for silent 'e'
        if word.endswith('e'):
            syllable_count -= 1
        
        # At least 1 syllable
        if syllable_count == 0:
            syllable_count = 1
            
        return syllable_count
    
    total_syllables = sum(count_syllables(word) for word in words)
    
    # Flesch-Kincaid Grade Level formula
    avg_words_per_sentence = total_words / total_sentences
    avg_syllables_per_word = total_syllables / total_words
    
    fk_grade = 0.39 * avg_words_per_sentence + 11.8 * avg_syllables_per_word - 15.59
    
    return {
        'label': label,
        'fk_grade_level': fk_grade,
        'avg_syllables_per_word': avg_syllables_per_word,
        'avg_words_per_sentence': avg_words_per_sentence
    }

print("✓ Readability Indices function loaded successfully")
print("  - calculate_flesch_kincaid(text, label)")
```

### CLASS 1 - READABILITY INDICES

```python
# Analyze readability for Class 1 novels
class1_readability_results = []

for novel_name, text in novels_class1:
    results = calculate_flesch_kincaid(text, label=f"{novel_name} (Class 1)")
    class1_readability_results.append(results)

class1_readability_df = pd.DataFrame(class1_readability_results)

print("="*70)
print("CLASS 1 - READABILITY INDICES")
print("="*70)
print(class1_readability_df.to_string(index=False))
print("="*70)
print(f"\nAverage FK Grade Level: {class1_readability_df['fk_grade_level'].mean():.2f}")
```
